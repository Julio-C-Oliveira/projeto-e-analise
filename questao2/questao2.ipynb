# Questão 2: Segunda Guerra Mundial


Seguindo as orientações dos generais:
- **General Edward:** O risco é diretamente proporcional à distância. Usaremos as distâncias mínimas como pesos do grafo.
- **General Jerzy:** A divisão sobressalente deve permanecer em seu respectivo aquartelamento (custo zero de deslocamento).

### Mapeamento dos Vértices do Grafo:
* `0`: $\alpha_1$ | `1`: $\alpha_2$ | `2`: $\alpha_3$ (Origens / Aquartelamentos)
* `3`: Nó Intermediário 1 | `4`: Nó Intermediário 2
* `5`: $\beta_1$ | `6`: $\beta_2$ | `7`: $\beta_3$ (Destinos / Frentes de Combate)

O primeiro passo do algoritmo é montar uma matriz de adjacenscias do grafo:

| **$\alpha_2$** | $\infty$ | 0 | $\infty$ | 2 | 3 | $\infty$ | $\infty$ | $\infty$ |
| **$\alpha_3$** | $\infty$ | $\infty$ | 0 | $\infty$ | 1 | $\infty$ | $\infty$ | 4 |
| **1** | $\infty$ | $\infty$ | $\infty$ | 0 | 1 | 2 | 4 | $\infty$ |

Para obter as distâncias mínimas de **todos para todos**, utilizaremos o algoritmo de **Floyd-Warshall** ($O(V^3)$), ideal para grafos pequenos e com múltiplas origens e destinos.

Antes de iniciar a execução é possível verificar que rodar o algoritmo para a matriz completa demandará instruções que serão redundantes. Todo $\alpha$ possui apenas vértices de saída e todo $\beta$ possui apenas vértices de entrada. Assim, é possível eliminar as respectivas colunas e linhas da matriz original. 


| Vértice               | **1** | **2** | **$\beta_1$** | **$\beta_2$** | **$\beta_3$** |
|:--------------|:-:|:-:|:---------:|:---------:|:---------:|
| **$\alpha_1$** | 1 | 3 | $\infty$ | $\infty$ | $\infty$ |
| **$\alpha_2$** | 2 | 3 | $\infty$ | $\infty$ | $\infty$ |
| **$\alpha_3$** | $\infty$ | 1 | $\infty$ | $\infty$ | 4 |
| **1**          | 0 | 1 | 2 | 4 | $\infty$ |
| **2**          | $\infty$ | 0 | 3 | 5 | 2 |


### Execução do Algoritmo de Floyd-Warshall

O algoritmo compara todos os caminhos possíveis através de loops aninhados estruturados, atualizando a distância se um nó intermediário `k` oferecer um atalho mais curto entre `i` e `j`:
$$D[i][j] = \min(D[i][j], D[i][k] + D[k][j])$$

Após a execução do algoritmo de Floyd-Warshall obtêm-se a seguinte matriz de adjacências:

| Vértice               | **1** | **2** | **$\beta_1$** | **$\beta_2$** | **$\beta_3$** |
|:--------------|:-:|:-:|:---------:|:---------:|:---------:|
| **$\alpha_1$** | 1 | 2 | 3 | 5 | 4 |
| **$\alpha_2$** | 2 | 3 | 4 | 6 | 5 |
| **$\alpha_3$** | $\infty$ | 1 | 4 | 6 | 3 |
| **1**          | 0 | 1 | 2 | 4 | 3 |
| **2**          | $\infty$ | 0 | 3 | 5 | 2 |

Com a matriz de adjacências posta é possível aplicar a técnica de programação linear para solucionar o problema.

Para isso temos:
$$\min Z = \sum_{i=1}^{3} \sum_{j=1}^{4} c_{ij} x_{ij}$$

Sendo $x_{ij}$ a quantidade de tropas enviadas do vértice $i$ até o vértice $j$ e $c_{ij}$ o custo desse trajeto dado pela matriz obtida. 
As restrições dessa simulação são:

$x_{1,1}$ + $x_{1,2}$ + $x_{1,3}$ + $x_{1,mantem}$ = 2

$x_{2,1}$ + $x_{2,2}$ + $x_{2,3}$ + $x_{2,mantem}$ = 2

$x_{3,1}$ + $x_{3,2}$ + $x_{3,3}$ + $x_{3,mantem}$ = 1 

Ou seja, $\alpha_1$ possui 2 tropas disponíveis, $\alpha_2$ possui 2 tropas disponíveis e $\alpha_3$ possui apenas uma tropa disponível.

Da mesma maneira têm-se:

$x_{1,1} + x_{2,1} + x_{3,1} = 2$

$x_{1,2} + x_{2,2} + x_{3,2} = 1$

$x_{1,3} + x_{2,3} + x_{3,3} = 1$

$x_{1,mantem} + x_{2,mantem} + x_{3,mantem} = 1$

Os limites da somatória representam a quantidade de origens ($\alpha_1$, $\alpha_2$ e $\alpha_3$) e a quantidade de destinos ($\beta_1$, $\beta_2$, $\beta_3$ e mantém, representando a tropa que permanecerá no quartel)

### Função Objetivo por Extenso

Expandindo o somatório da função objetivo $\min Z = \sum_{i=1}^{3} \sum_{j=1}^{4} c_{ij} x_{ij}$ com os custos mínimos obtidos pela matriz do algoritmo de Floyd-Warshall, temos a seguinte expressão:

$$
\min Z = 3x_{1,1} + 5x_{1,2} + 4x_{1,3} + 0x_{1,mantem} + 4x_{2,1} + 6x_{2,2} + 5x_{2,3} + 0x_{2,mantem} + 4x_{3,1} + 6x_{3,2} + 3x_{3,3} + 0x_{3,mantem}
$$

### Raciocínio para a Distribuição das Tropas

Para encontrar os valores de cada $x_{ij}$ de forma lógica, analisamos o custo de atendimento de cada destino (frente de combate), priorizando sempre as rotas mais econômicas:

#### 1. Atendimento da Frente $\beta_3$ (Demanda: 1 tropa)
Analisando os custos de deslocamento de todas as origens até $\beta_3$:
* **$\alpha_1$ para $\beta_3$:** custo 4
* **$\alpha_2$ para $\beta_3$:** custo 5
* **$\alpha_3$ para $\beta_3$:** custo 3

> **Decisão:** A rota mais barata é enviar a única tropa disponível de **$\alpha_3$ para $\beta_3$** ($x_{3,3} = 1$). Com isso, a frente $\beta_3$ é totalmente suprida e o quartel $\alpha_3$ encerra suas atividades.

#### 2. Atendimento da Frente $\beta_1$ (Demanda: 2 tropas)
Restam as origens $\alpha_1$ (2 tropas) e $\alpha_2$ (2 tropas). Os custos para a linha de frente $\beta_1$ são:
* **$\alpha_1$ para $\beta_1$:** custo 3
* **$\alpha_2$ para $\beta_1$:** custo 4

> **Decisão:** Para preencher as 2 vagas de $\beta_1$, pegamos **1 tropa de $\alpha_1$** ($x_{1,1} = 1$) e **1 tropa de $\alpha_2$** ($x_{2,1} = 1$).

#### 3. Atendimento da Frente $\beta_2$ (Demanda: 1 tropa) e Retenção (`mantem`)
Agora resta apenas 1 tropa disponível em $\alpha_1$ e 1 tropa em $\alpha_2$.
* O custo para atender $\beta_2$ vindo de $\alpha_1$ é **5**, enquanto vindo de $\alpha_2$ é **6**.
* O custo para manter a tropa no próprio quartel é **0** para ambas as origens.

> **Decisão:** Como $\alpha_1$ chega mais barato em $\beta_2$ do que $\alpha_2$, enviamos a última tropa de **$\alpha_1$ para $\beta_2$** ($x_{1,2} = 1$). Consequentemente, a tropa restante de **$\alpha_2$ permanece no aquartelamento** ($x_{2,mantem} = 1$) com custo zero.

### Substituindo valores


Ao resolver o modelo de programação linear considerando as restrições de oferta e demanda do problema, encontramos a seguinte alocação ótima para as variáveis de decisão $x_{ij}$:

* **Origem $\alpha_1$:** $x_{1,1} = 1$, $x_{1,2} = 1$, $x_{1,3} = 0$, $x_{1,mantem} = 0$
* **Origem $\alpha_2$:** $x_{2,1} = 1$, $x_{2,2} = 0$, $x_{2,3} = 0$, $x_{2,mantem} = 1$
* **Origem $\alpha_3$:** $x_{3,1} = 0$, $x_{3,2} = 0$, $x_{3,3} = 1$, $x_{3,mantem} = 0$

Substituindo esses valores numéricos diretamente na equação por extenso, obtemos o custo total do deslocamento:

$$
Z = 3(1) + 5(1) + 4(0) + 0(0) + 4(1) + 6(0) + 5(0) + 0(1) + 4(0) + 6(0) + 3(1) + 0(0)
$$

$$
Z = 3 + 5 + 4 + 3 = 15
$$

O valor do custo mínimo total de deslocamento das tropas para as frentes de combate é **15**.

### 1. Modelagem do Problema

O problema foi modelado utilizando uma abordagem híbrida de **Teoria dos Grafos** e **Programação Linear (Problema de Transporte)**, estruturada em duas camadas principais:

* **Infraestrutura de Transporte (Grafo Dirigido):** O campo de operações foi representado como um dígrafo $G = (V, E)$, onde os vértices ($V$) englobam os pontos de origem ($\alpha_i$), os nós de passagem/rotatórias intermediárias e os destinos finais de combate ($\beta_j$). As arestas ($E$) possuem pesos ponderados que equivalem à distância física de deslocamento, a qual é diretamente proporcional ao risco da operação militar.
* **Modelo de Otimização (Programação Linear):** O cenário foi mapeado como um problema clássico de transporte balanceado. Onde as variáveis de decisão $x_{ij}$ mensuram a quantidade física de contingente alocado da origem $i$ para o destino $j$. O modelo busca minimizar a função de custo global:

$$\min Z = \sum_{i=1}^{3} \sum_{j=1}^{4} c_{ij} x_{ij}$$

Sujeito a restrições estritas de **Capacidade de Oferta** nos aquartelamentos (garantindo que nenhuma origem envie mais tropas do que possui disponível) e **Garantia de Demanda** nas linhas de frente (assegurando o suprimento estratégico exato exigido por cada fronte de batalha, incluindo o nó fictício de retenção logística `mantem`).

### 2. Análise dos Algoritmos e Complexidade Computacional

A resolução do problema seguiu uma esteira de execução dividida em duas etapas sequenciais, cada uma possuindo seu respectivo impacto computacional:

#### Etapa 1: Algoritmo de Floyd-Warshall
Utilizado para calcular o caminho mínimo entre todos os pares de nós ($All-Pairs\ Shortest\ Path$), consolidando a matriz de custos logísticos de ponta a ponta ($c_{ij}$).
* **Complexidade Teórica:** $\mathcal{O}(V^3)$, onde $V$ é o número total de vértices.
* **Complexidade Aplicada:** Como o grafo possui estrutura bipartida oculta (as origens possuem apenas grau de saída e os destinos apenas grau de entrada), realizou-se uma otimização prévia reduzindo a dimensão da matriz operacional. Para o cenário com 8 vértices originais, a execução executa pouquíssimas iterações, tornando-se extremamente eficiente.

#### Etapa 2: Resolução do Modelo Linear (Simplex / Métodos de Pontos Interiores)
A resolução das equações e inequações geradas pela distribuição de tropas é feita classicamente via algoritmo **Simplex** ou através de solvers dedicados (como CPLEX, GLPK ou Gurobi).
* **Complexidade Teórica:** O Simplex possui complexidade no pior caso de tempo exponencial $\mathcal{O}(2^n)$, mas demonstra desempenho prático polinomial na esmagadora maioria dos casos reais. 
* **Complexidade Aplicada:** Para problemas estruturados de transporte com $M$ origens e $N$ destinos, algoritmos especializados como o *Network Simplex* resolvem a alocação de forma fortemente polinomial, operando de maneira extremamente ágil para a matriz reduzida $3 \times 4$.

### 3. Conclusão e Resultados Obtidos

A estratégia de otimização matemática provou ser imperativa para o planejamento logístico militar, mitigando caminhos redundantes e eliminando o processo de tentativa e erro. A filtragem prévia topológica das colunas e linhas da matriz original acelerou a convergência do algoritmo de caminhos mínimos, permitindo que a Programação Linear operasse apenas com dados úteis.

A análise lógica aplicada sobre a matriz de distâncias mínimas confirmou que a melhor distribuição de contingente atende prioritariamente às frentes mais isoladas ou de acesso restrito (como a cadeia de suprimento de $\alpha_3 \rightarrow \beta_3$), deixando margem de manobra para que os quartéis com maior mobilidade atendam às demandas remanescentes.

**Resumo Econômico da Operação:**
* **Custo Total de Deslocamento ($Z$):** **15 unidades de custo/risco.**
* **Eficiência Estratégica:** 100% das demandas de combate foram supridas dentro dos limites exatos de contingente disponível, garantindo o custo mínimo global e mantendo 1 tropa sobressalente em segurança na base de origem $\alpha_2$ com custo zero de movimentação.
